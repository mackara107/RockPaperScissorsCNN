Loading all of the rock, paper, and scissor images into lists loaded_rock, loaded_paper, loaded_scissor. These will be useful because now we can access all the groups of images at once to edit them.

Importing image and pyplot from matplotlib to show 3 example images, one of each category, in our dataset. Then, below the images are the numpy arrays representing each image that is stored inside each of our lists.

Converting all images to greyscale to avoid any confusion with coloring of skin. To do this, importing numpy and setting rgb weight which will remove all color from image. We have to use np.dot(), which takes the array and the new rgb values, to do the transformation. Then, appending all gray images to a new list called final_rock, final_paper, final_scissor. The images that were displayed above are now shown in their final grayscale form below.

```python
import numpy as np
rgb_weights = [0.2989, 0.5870, 0.1140]

final_rock = list()
for img in loaded_rock:
    grayscale_image = np.dot(img[...,:3], rgb_weights)
    final_rock.append(grayscale_image)
    
final_paper = list()
for img in loaded_paper:
    grayscale_image = np.dot(img[...,:3], rgb_weights)
    final_paper.append(grayscale_image)
    
final_scissor = list()
for img in loaded_scissor:
    grayscale_image = np.dot(img[...,:3], rgb_weights)
    final_scissor.append(grayscale_image)

pyplot.subplot(311)
pyplot.imshow(final_rock[0], cmap=pyplot.get_cmap("gray"))
pyplot.subplot(312)
pyplot.imshow(final_paper[0], cmap=pyplot.get_cmap("gray"))
pyplot.subplot(313)
pyplot.imshow(final_scissor[0], cmap=pyplot.get_cmap("gray"))
pyplot.show()
```

# Labeling Data
Now we will need to grab our data and assign all images into one of the three classes: Rock, Paper, or Scissor.

To do this, the images are taken from my directory. We also resize them to 32 X 32 images using target_size=(32,32). Then, we explicitly specify the classes in the order they appear in folders in my directore. We also clarify that we need categorical labels since we use categorical_crossentropy loss. 

We will be using ImageDataGenerator. Keras’s ImageDataGenerator is used to perform image augmentation by directly reading the CSV files through pandas dataframe.
Most often the Image datasets available on the internet are either has images placed under folders which has their respective class names or placed under a single folder along with a CSV or JSON file which maps the image filenames with their corresponding classes.

We will be using the flow_from_directory method that helps you read the images from the folders. We specify the directory which contains the three folders for the three classes, and then specify the class names.

```python
batch_size = 128
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1/255)

train_generator = train_datagen.flow_from_directory(
        'C:/Users/kara/Desktop/Compsci/AI/rockpaperscissor/rps-cv-images', 
        target_size=(32, 32),
        batch_size=batch_size,
       
        classes = ['paper','rock','scissors'],
        
        class_mode='categorical')
```

This output shows that all of our 2188 images were separated successfully into their appropriate of the three classes.

# Creating the Model

Now we have to build the model. First, we input the necessary functions from keras.

```python
from tensorflow.python.keras.layers import Dense, Conv2D, Flatten, MaxPooling2D
from tensorflow.python.keras import Sequential
from tensorflow.python.keras.layers import Dropout
```

The model type that we will be using is Sequential. Sequential is the easiest way to build a model in Keras. It allows you to build a model layer by layer. We use the ‘add()’ function to add layers to our model. Our first layer is a Conv2D layer. These are convolution layers that will deal with our input images, which are seen as 2-dimensional matrices. 32 is the number of nodes in the layer.

Kernel size is the size of the filter matrix for our convolution. So a kernel size of 3 means we will have a 3x3 filter matrix.

Activation is the activation function for the layer. The activation function we will be using for our first layer is the ReLU, or Rectified Linear Activation.

The MaxPooling2D layer reduces the dimensionality (and thus required computational power) by outputting the maximum of the part of the input image that is captured by the filter matrix. In this case our MaxPool layer has a size of 2X2.

Our first layer also takes in an input shape. This is the shape of each input image, 28,28,1 as seen earlier on, with the 1 signifying that the images are greyscale.

In between the Conv2D layers and the dense layer, there is a ‘Flatten’ layer. Flatten serves as a connection between the convolution and dense layers.

‘Dense’ is the layer type we will use in for our output layer. We will have 10 nodes in our output layer, one for each possible outcome (0–9). The activation is ‘softmax’. Softmax makes the output sum up to 1 so the output can be interpreted as probabilities. The model will then make its prediction based on which option has the highest probability.

Dropout is a way to regularize a neural network and reduce overfitting. For this model, we want to be dropping out at 20% (Dropout(0.2)). This means that we will be retaining 80% of neurons.

```python
model = Sequential()

model.add(Conv2D(32, (3,3), activation="relu", input_shape=(32,32,3)))
model.add(Dropout(0.2))

model.add(Conv2D(32, (3,3), activation="relu"))
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Conv2D(64, (3,3), activation="relu"))
model.add(Dropout(0.2))

model.add(Conv2D(64, (3,3), activation="relu"))
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Conv2D(128, (3,3), activation="relu", padding="same"))
model.add(Dropout(0.2))

model.add(Conv2D(128, (3,3), activation="relu", padding="same"))
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Flatten())

model.add(Dropout(0.2))
model.add(Dense(1024, activation="relu"))

model.add(Dropout(0.2))
model.add(Dense(512, activation="relu"))

model.add(Dropout(0.2))
model.add(Dense(3, activation="softmax"))
```

To get a summary of the details of our model, we will use model.summary() which shows all layers and their output shapes.

```python
model.summary()
```

Now we will compile our model with the line below. 

loss='categorical_crossentropy'. The loss function sorts outputs into categories.

RMSprop optimizes the model

metrics=['accuracy'] Lastly, we want to track accuracy on top of the loss function.

```python
from tensorflow.keras.optimizers import RMSprop

model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(lr=0.001),
              metrics=['acc'])
```

Below is the total number of samples found, saved into total_sample.

```python
total_sample=train_generator.n
```

Now, we will fit our model to the data. We are using train_generator, which contains all of the data in their 3 classes. Total_sample / batch_size is the total size used for the steps for each epoch. To get a good accuracy, we will do 30 epochs.

```python
history = model.fit_generator(
        train_generator, 
        steps_per_epoch=int(total_sample/batch_size),  
        epochs=30,
        verbose=1)
```

From tracking the progress through our 30 epochs, it appears that we were successful in getting an accuracy close to 1 and a loss close to 0. To visualize it, we will pyplot to plot our accuracy and loss over the course of the training.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,4))
plt.plot([i+1 for i in range(30)],history.history['acc'],'-o',c='k',lw=2,markersize=9)
plt.grid(True)
plt.title("Training accuracy with epochs\n",fontsize=18)
plt.xlabel("Training epochs",fontsize=15)
plt.ylabel("Training accuracy",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.plot([i+1 for i in range(30)],history.history['loss'],'-o',c='k',lw=2,markersize=9)
plt.grid(True)
plt.title("Training loss with epochs\n",fontsize=18)
plt.xlabel("Training epochs",fontsize=15)
plt.ylabel("Training loss",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

Our graphs do show that our loss and accuracy achieved the goal state. Now that we have confirmed our model is ideal, we can test our own pictures on our model and see if it predicts correctly.

# Game Feature
Now that we have our model working, we can implement a basic game feature that will predict the winner of Rock-Paper-Scissors!
 
The basic rules:
- PAPER beats ROCK ([0] beats [1])
- ROCK beats SCISSOR ([1] beats [2])
- SCISSOR beats PAPER ([2] beats [0])
- TIE if both are the same

```python
def checkWinner(player1,player2):
    p1 = player1[0] #int inside the array
    p2 = player2[0]
    classes = ["Paper","Rock","Scissors"]
    
    if(p1 == p2):
        print("Game is a tie")
    if((p1==0 and p2==1)or(p1==1 and p2==2)or(p1==2 and p2==0)):
        print("Player 1 Wins with",classes[p1])
    if((p2==0 and p1==1)or(p2==1 and p1==2)or(p2==2 and p1==0)):
        print("Player 2 Wins with",classes[p1])
```

The checkWinner function takes the output from model.predict_classes() of two different images and will print out who the winner is. For example, we will use our test1 and test2 as parameters and see the output

```python
checkWinner(test1,test2)
```

We will test with all possible options to make sure the method works (Excluding tie options).

```python
checkWinner([0],[1])
checkWinner([1],[2])
checkWinner([2],[0])
checkWinner([1],[0])
checkWinner([2],[1])
checkWinner([0],[2])
```
